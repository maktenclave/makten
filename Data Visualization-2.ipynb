```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([3,8,1,10,5,7])
print(plt.plot(y,linestyle='dashed'))
plt.title('Sample graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
print(plt.show())
```

```python
y=np.array([3,8,1,10,7])
print(plt.plot(y,marker='o'))
print(plt.show())
```

```python
x=np.array([0,6])
y=np.array([0,250])
print(plt.plot(x,y,'violet'))
print(plt.show())
```

```python
print(plt.plot(x,y,linewidth='30',color='c'))
print(plt.show())
```

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
print(plt.plot(y1))
print(plt.plot(y2))
plt.grid()
print(plt.show())
```

```python
x=np.array([0,1,2,3,4,5])
y=np.array([3,8,1,10,9,4])
plt.subplot(1,2,2)
plt.plot(x,y)
```

```python
x=np.array([1, 2, 6, 8])
y=np.array([3, 8, 1, 10])
plt.subplot(1,2,2)
plt.scatter(x, y)
plt.show()
```
